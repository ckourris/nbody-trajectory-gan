```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Load Data

```python

df = pd.read_csv("initial_conditions/index.csv")

# quick summary
print(df.describe())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# N distribution
df["N"].value_counts().sort_index().plot(
    kind="bar", ax=axes[0], color="steelblue"
)
axes[0].set_title("Number of bodies")
axes[0].set_xlabel("N")


# virial ratio
sns.histplot(df["virial_ratio"], bins=20, ax=axes[1], color="steelblue")
axes[1].axvline(0.5, color="red", linestyle="--", label="Q=0.5")
axes[2].axvline(1.0, color="orange", linestyle="--", label="Q=1.0")
axes[1].set_title("Virial ratio distribution")
axes[1].legend()

# mass ratio (log scale)
sns.histplot(np.log10(df["mass_ratio_max"]), bins=20, ax=axes[2], color="steelblue")
axes[2].set_title("Mass ratio max (log10)")
axes[2].set_xlabel("log10(max/min mass)")

plt.tight_layout()
```
